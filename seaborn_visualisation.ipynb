```python
# Distribution plot
# This helps to know how well your data is distributed. If it is normal
#right or left skewd, this can help you carry out staistical analysis on the dat  
sns.distplot(iris_table['sepal_length'])
```

```python
# Join plot 
sns.jointplot(x='sepal_length', y='sepal_width', data=iris_table)
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris_table)
```

```python
sns.violinplot(x="species",y="sepal_length",data=iris_table)
```

```python
sns.pairplot(data = iris_table, kind = 'scatter', hue='species')
```

```python
sns.heatmap(data = iris_table.corr(),annot = True)
```
